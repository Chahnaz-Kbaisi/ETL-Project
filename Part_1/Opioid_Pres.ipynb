```python
# Import Dependencies:
import pandas as pd
import os
import numpy as np
```

```python
Prescription_VA = os.path.join("Resources", "Opioid_Pres_VA.csv")
Prescription_Retail = os.path.join("Resouces", "Opioid_Retail.csv")

```

```python
# Reading in data file to store into Pandas DataFrame:
Prescription_VA_df = pd.read_csv("Resources/Opioid_Pres_VA.csv")
Prescription_VA_df.head()
```

```python
Prescription_VA_df['Percent Change'] = Prescription_VA_df['Percent Change'].astype(str)
Prescription_VA_df
```

```python
# Cleaning dataset and dropping any bad records:
cleaned_Prescription_VA_df = Prescription_VA_df.dropna(how='any')
cleaned_Prescription_VA_df.head()
```

```python
# Filtering dataset by selecting columns needed to answer potential query:
Prescription_VA_subset = cleaned_Prescription_VA_df[["Percent Change (integer)", "city", "2012 Prescribing Rate", "2018 Prescribing Rate", "Non-VA Prescribing Rate (Regional Use)", "state_full", "state_postal"]]

Prescription_VA_subset.head()
```

```python
# Renaming the subset dataframe columns:
Prescription_rate = Prescription_VA_subset.rename(columns={
    'Percent Change (integer)': 'Percent Change',
    'Non-VA Prescribing Rate (Regional Use)': 'Non-VA Prescribing Rate'})
Prescription_rate.head()
```

```python
# Saving to csv: 
Prescription_rate.to_csv('Resources/Prescription_rate.csv', index=False)
```
